```python
def calculate_insurance_cost(age, bmi, children, smoker):
    """
    Calculate estimated insurance cost based on simple formula.
    Formula is illustrative, not real-world accurate.
    """
    base_cost = 250  # base monthly premium in USD
    age_factor = (age - 18) * 5 if age > 18 else 0
    bmi_factor = (bmi - 21) * 10 if bmi > 21 else 0
    children_factor = children * 20
    smoker_factor = 150 if smoker else 0
```

```python
# -------------------------------
# Example Data

ages = list(range(18, 66, 5))  # Ages from 18 to 65
bmi = 26                       # Fixed BMI for this example
children = 2                   # Fixed number of children
smoker = False                 # Non-smoker
```

```python
# Calculate costs for each age
costs = [calculate_insurance_cost(age, bmi, children, smoker) for age in ages]
```

```python

# Visualization with Matplotlib

plt.figure(figsize=(8, 5))
plt.plot(ages, costs, marker='o', color='blue', label='Insurance Cost')
plt.title("Estimated Health Insurance Cost by Age")
plt.xlabel("Age (years)")
plt.ylabel("Monthly Premium (USD)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Load data
age_charges = defaultdict(list)
```

```python
# Calculate average charges per age
ages = sorted(age_charges.keys())
avg_charges = [sum(age_charges[a]) / len(age_charges[a]) for a in ages]
```

```python
# Plot
plt.plot(ages, avg_charges, marker="o", color="blue")
plt.xlabel("Age")
plt.ylabel("Average Charges ($)")
plt.title("Average Insurance Charges by Age")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
# Copy code
import matplotlib.pyplot as plt
```

```python
# Example claim data (in USD)
auto_claims = [1200, 800, 1500, 2000]   # Auto insurance claims
health_claims = [5000, 3000, 4500, 4000]  # Health insurance claims
```

```python
# 1️⃣ Calculate total amounts
total_auto = sum(auto_claims)
total_health = sum(health_claims)
```

```python
# 2️⃣ Prepare data for plotting
categories = ['Auto Insurance', 'Health Insurance']
totals = [total_auto, total_health]
```

```python
# Create bar chart
plt.figure(figsize=(6, 4))
bars = plt.bar(categories, totals, color=['skyblue', 'lightgreen'])
```

```python
# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 100, f"${height:,}", 
             ha='center', va='bottom', fontsize=10)
```

```python
# Show plot
plt.tight_layout()
plt.show()
```
